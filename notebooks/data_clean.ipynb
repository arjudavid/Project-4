```python
# Import required libraries
import os
import pandas as pd
```

```python
# Set Kaggle API key directory
os.environ["KAGGLE_CONFIG_DIR"] = "../kaggle_key"
```

```python
# Import Kaggle API and authenticate
from kaggle.api.kaggle_api_extended import KaggleApi
api = KaggleApi()
api.authenticate()
```

```python
# Download and extract dataset to the specified folder
dataset_path = "../data"
api.dataset_download_files("arianazmoudeh/airbnbopendata", path=dataset_path, unzip=True)
```

```python
# Load the downloaded dataset into a DataFrame
file_path = "../data/airbnb_open_data.csv"
df = pd.read_csv(file_path)
```

```python
# Display the DataFrame
df.head()
```

```python
# Drop specific columns from the DataFrame
drop_columns = [
'id', 'host id', 'host name',
'country', 'country code', 'last review',
'house_rules', 'license'
]

df_cln = df.drop(columns=drop_columns)
df_cln.head()
```

```python
drop_column = 'NAME'

df_cln = df_cln.drop(columns=drop_column)
df_cln.head()
```

```python
# Remove rows with null or empty values 
df_cln = df_cln.dropna(subset=[
    'host_identity_verified', 'neighbourhood group',
    'neighbourhood', 'lat', 'long', 'instant_bookable',
    'cancellation_policy', 'room type', 'Construction year', 'price'
])

df_cln.head()
```

```python
# Remove rows where 'minimum nights' or 'availability 365' is less than 0
df_cln = df_cln[(df_cln['minimum nights'] >= 0) & (df_cln['availability 365'] >= 0)]

df_cln.head()
```

```python
# Display data types for all columns
df_cln.dtypes
```

```python
# Convert 'host_identity_verified' and 'instant_bookable' to category
category_columns = ['host_identity_verified', 'instant_bookable']

for col in category_columns:
    df_cln[col] = df_cln[col].astype('category')

# Verify the change
print(df_cln.dtypes[category_columns])
```

```python
# Fill NaN values with 0 in specified columns
columns_to_fill = [
    'Construction year', 'price', 'service fee', 
    'minimum nights', 'number of reviews', 
    'reviews per month', 'review rate number', 
    'calculated host listings count', 'availability 365'
]

df_cln[columns_to_fill] = df_cln[columns_to_fill].fillna(0)
df_cln.head()
```

```python
# Clean and standardize 'neighbourhood group'
df_cln['neighbourhood group'] = df_cln['neighbourhood group'].str.strip().str.lower()

# Standardize values with corrections
standardized_neighbourhood_group = {
    'brooklyn': 'Brooklyn',
    'brookln': 'Brooklyn',
    'manhattan': 'Manhattan',
    'manhatan': 'Manhattan',
    'queens': 'Queens',
    'staten island': 'Staten Island',
    'bronx': 'Bronx'
}

df_cln['neighbourhood group'] = df_cln['neighbourhood group'].map(standardized_neighbourhood_group)
```

```python
# Standardize 'neighbourhood' values
neighbourhood_corrections = {
    "Hell'S Kitchen": "Hell's Kitchen",
    "Prince'S Bay": "Prince's Bay",
    "Bull'S Head": "Bull's Head",
    "Bay Terrace, Staten Island": "Bay Terrace",
    "Chelsea, Staten Island": "Chelsea",
    "Columbia St": "Columbia Street",
    "Soho": "SoHo"
}

df_cln['neighbourhood'] = df_cln['neighbourhood'].replace(neighbourhood_corrections)
```

```python
# Define a reusable function to clean and convert columns
def clean_and_convert(column):
    return pd.to_numeric(
        df[column]
        .astype(str)
        .str.strip()
        .replace({'': None, 'N/A': None})
        .str.replace(r'[^\d.]', '', regex=True),
        errors='coerce'
    )

# Clean and convert the 'price' and 'service fee' columns
df_cln['price'] = clean_and_convert('price')
df_cln['service fee'] = clean_and_convert('service fee')
```

```python
# List of columns to clean and convert to integers
columns_to_convert = [
    'price', 'service fee', 'Construction year', 
    'minimum nights', 'availability 365', 
    'number of reviews', 'calculated host listings count'
]

# Fill NaN values with 0 and convert columns to integer type
df_cln[columns_to_convert] = df_cln[columns_to_convert].fillna(0).astype(int)
```

```python
# Display the number of non-null records per column
print(df_cln.count())
```

```python
# Save csv copy in data folder
df_cln.to_csv('../data/airbnb_cleaned.csv', index=False)
```
